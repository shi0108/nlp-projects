```python
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
from sklearn.metrics import mean_squared_error
```

```python
# preprocess data
close_price = df['CLOSE'].values.reshape(-1, 1)
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_close_price = scaler.fit_transform(close_price)
```

```python
# Split the data
train_size = int(len(scaled_close_price) * 0.8)
train_data = scaled_close_price[:train_size]
test_data = scaled_close_price[train_size:]
```

```python

  #  Create a Time Series Dataset
  def create_dataset (df, time_step=60):
    x, y = [], []
    for i in range(len(df) - time_step - 1):
        a = df[i:(i + time_step), 0]
        x.append(a)
        y.append(df[i + time_step, 0])
    return np.array(x), np.array(y)
```

```python
# Reshape data
time_step = 60
x_train, y_train = create_dataset(train_data, time_step)
x_test, y_test = create_dataset(test_data, time_step)
```

```python
# Build the RNN Model
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(time_step, 1)))
```

```python
model.add(LSTM(units=50, return_sequences=False))
```

```python
model.add(Dense(units=1))
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
# train thr model
model.fit(x_train, y_train, batch_size=32, epochs=100)

```
